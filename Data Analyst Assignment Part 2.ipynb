```python
import pandas as pd

new = pd.read_csv(r'C:\Users\sidkr\OneDrive - RG Insurance Processing Services Private Limited\Desktop\Assignment For DA Candidates\Assignment For DA Candidates\Data Analyst Assignment Part 2\bigkart_newsales.csv')
old = pd.read_csv(r'C:\Users\sidkr\OneDrive - RG Insurance Processing Services Private Limited\Desktop\Assignment For DA Candidates\Assignment For DA Candidates\Data Analyst Assignment Part 2\bigkart_sales.csv')
```

```python
# Question 1: Some of the orders are stored in another csv file named `bigkart_newsales`. Read the csv file, store it in a data frame and add it to the `bigkart_sales` data frame. Find the total sales value of the category 'Office Supplies' after combining the dataframes

data = pd.concat([new, old], axis = 0)
data = data.reset_index()
data = data.drop(['index'], axis = 1)
total_sales_office_supplies = sum(data.Sales[data.Category=='Office Supplies'])
total_sales_office_supplies
```

```python
# Question 2: There are some duplicate rows in the data frame. Drop these rows and calculate the total sales value of the category Office Supplies.

data = data.drop_duplicates()
total_sales_office_supplies = sum(data.Sales[data.Category=='Office Supplies'])
total_sales_office_supplies
```

```python
# Question 3: Find the most profitable category and sub category combination based on the net profit.

part_3 = pd.DataFrame()
part_3['Category + Sub Category'] = data['Category']+' '+data['Sub-Category']
part_3['Profit'] = data['Profit']
temp_df_categories_profit = pd.DataFrame(part_3[['Profit', 'Category + Sub Category']].groupby('Category + Sub Category')['Profit'].sum())
temp_df_categories_profit = temp_df_categories_profit.sort_values('Profit', axis = 0, ascending = False).reset_index()
best_combination = temp_df_categories_profit['Category + Sub Category'][0]
best_combination
```

```python
# Question 4: How many invalid order IDs are there in the data frame. An order id is of the form AZ-2011-Y where Y represents a whole number. A Order ID is said to be valid only if Y consists of 7 digits. Find the number of invalid order order IDs in the data frame.

number_invalid_order_ids = len([i for i in data['Order ID'] if len(i)<15])
number_invalid_order_ids
```

```python
# Question 5: Find the top 25 orders based on sales value and find the number of orders which belong to furniture category.

top_25_orders_sales = data.sort_values("Sales", axis = 0, ascending = False).iloc[:25, :]
top_25_orders_sales
number_orders_furniture_category = len(data[data.Category == 'Furniture'])
number_orders_furniture_category
```

```python
# Question 6: Among the orders with sales>250 and profit>50, find the product name of the fourth highest order based on sales value.

predict_4th_highest_sales = data[(data.Sales>250) & (data.Profit>50)].sort_values("Sales", axis = 0, ascending = False).iloc[3, :]['Product Name']
predict_4th_highest_sales
```

```python
# Question 7: Remove the orders with negative profit by dropping the corresponding rows with negative `Profit`. Find the product that makes the lowest profit per Quantity in the Technology category.

part_7 = data[~data.Profit<0]
part_7 = part_7[data.Category == 'Technology']
part_7['Profit per Quantity'] = round(part_7['Profit']/part_7['Quantity'], 2)
part_7 = part_7.sort_values('Profit per Quantity', axis = 0, ascending = True)
lowest_profit_product_tech = part_7.iloc[:1, :]
lowest_profit_product_tech

# there is no profit in selling "Brother Ink, Laser". So if we remove this product considering that 0 profit might be an 
# erroneous data entry, then the next product that sold for the least profit per quantity is: 
# 'Logitech Keyboard, Programmable': part_7.iloc[1:2, :]
```
